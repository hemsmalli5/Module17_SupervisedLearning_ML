```python
import pandas as pd
from path import Path

file_path = Path("../Resources/loans_data.csv")
loans_df = pd.read_csv(file_path)
loans_df.head()
```

```python
# As part of the preprocessing steps, perform encoding; convert test features into machine readable numbers.

# we will use pandas pd.get_dummies method here

loans_binary_encoded = pd.get_dummies(loans_df, columns=["gender"])
loans_binary_encoded.head()
```

```python
# encode multiple columns at the same time.

loans_binary_encoded = pd.get_dummies(loans_df, columns=["education", "gender"])
loans_binary_encoded.head()
```

```python
# Scikit-learn’s LabelEncoder module can also transform text into numerical data.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = loans_df.copy()
df2['education'] = le.fit_transform(df2['education'])
df2['gender'] = le.fit_transform(df2['gender'])
df2
```

```python
# Month encoding doesn't give expected results. lets try with new instance of label encoder
label_encoder = LabelEncoder()
loans_df["month_le"] = label_encoder.fit_transform(loans_df["month"])
loans_df.head()
```

```python
# Above August was referred as 1. Hence lets try different method
# Create a dictionary of months of the year and apply custom function to convert months names to corresponding integers:

months_num = {
    "January":1,
    "February":2,
    "March":3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}
```

```python
loans_df["month_num"] = loans_df["month"].apply(lambda x: months_num[x])
loans_df
```

```python
loans_df = loans_df.drop(["month", "month_le"], axis=1)
loans_df.head()
```
